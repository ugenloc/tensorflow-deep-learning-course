### Getting infomation from tensors**

Whwn dealing with tensors we want to be aware of the following attributes

* Shape: This is  the number of element of each of the dimension of a tensor

* Rank: This is the number of a tensors dimension. A scalar ranks 0, a vector has a ranks 1, a matrix has rank 2 and a tensor has rank n.

* Axis:A particular dimension of a tensor

* Size: The total number of item in a tensor.

```python
# create a rank 4 (4 dimension)
rank_4_tensor = tf.zeros(shape=([2, 3, 4, 5]))
rank_4_tensor

#_____ Breakdown
# 2 ->dmiensions
# 3 ->matrix in each dimension
# 4 ->rows in each matrix
# 5 ->column in each matrix
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor),
```

```python
## Get various attribute of our tensors
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimension (rank):", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along the 0 axis:", rank_4_tensor.shape[0])
print("Elements along the last axis:", rank_4_tensor.shape[-1])
print("Total number of elements in our tensor:", tf.size(rank_4_tensor).numpy())
```

**Matrix multiplication**

In machine learning, matrix multiplication is one of the most common tessor operations.

There are two rules our tensors (or matrices) need to fufil if we are going to matrix multiply them:

1. The inner dimension must match
2. The resulting matrix must have the outer dimension

```python
# Matrix multltiplication in tensorflow
tf.matmul(tensor, tensor)
```

```python
tensor_1 = tf.constant([[[[1, 2, 5],
                          [7, 2, 1],
                          [3, 3, 3]]]])
tensor_2 = tf.constant([[[[3, 5],
                          [6, 7],
                          [1, 8]]]])
tf.matmul(tensor_1, tensor_2)
```

```python
# matrix multiplication with python operator
tensor @ tensor
```

```python
tensor.shape
```

```python
# create a tensor of 3 rows and 2 columns
X =  tf.constant([[1, 2],
                   [3, 4],
                   [5, 6]])
# Create another tensor of (3,2)
Y = tf.constant([[7, 8],
                   [9, 10],
                   [11, 12]])
X, Y
Y.shape
```

```python
# Matrix multiply tensors of the same shape
tf.matmul(X, Y)
```

```python
# Try matrix multiplication with transpose rather than reshape
tf.matmul(tf.transpose(X), Y)
```

### Change the data type of a tensor

```python
# Create a new tensor with default datatype (float32)
B = tf.constant([1.7, 7.4])
B.dtype
```

```python
C = tf.constant([7, 10])
C.dtype
```

```python
# change from float 32 to float 16 (reduced precision)
D = tf.cast(B, dtype=tf.float16)
D, D.dtype
```

```python
# Change from int32 to float32
E = tf.cast(C, dtype=tf.float32)
E
C.dtype
```

```python
# If you ever find yourself stuckaccross the need to change a data type because some of the calculations aren't working correctly, tfcast() is your friend.
E_float16 = tf.cast(E, dtype=tf.float16)
E_float16
```
